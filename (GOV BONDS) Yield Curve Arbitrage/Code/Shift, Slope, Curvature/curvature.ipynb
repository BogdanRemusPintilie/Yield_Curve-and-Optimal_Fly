```python
import pandas as pd
import numpy as np

df = pd.read_excel("C:\\Users\\Bogdan\\OneDrive - University of Warwick\\Desktop\\Projects\\(GOV BONDS) Yield Curve Arbitrage\\Data\\Yields\\Yield.xlsx")

df['Date'] = pd.to_datetime(df['Date'])
jun_24_data = df[df['Date'] == '2024-06-01']

maturities = np.array([1/12, 2/12, 3/12, 4/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30])
yields = np.array(jun_24_data.iloc[0, 1:])

curvatures = np.zeros(len(maturities))#inititlise a 0 list that you update
        
for i in range(1, len(maturities) - 1):#need the -1 cause of the index transition that is 0123...
    # Previous and next maturity and yield
    prev_maturity= maturities[i-1]
    next_maturity = maturities[i+1]
    prev_yield = yields[i-1]
    next_yield = yields[i+1]
            
    # Linear interpolation to find the expected yield at the current maturity
    interp_yield = (next_yield - prev_yield) / (next_maturity - prev_maturity) * (maturities[i] - prev_maturity) + prev_yield #slope * change in x axis -> change in y axis (yield change), add to that the yield at the prev point
            
    # Curvature calculation
    curvatures[i] = yields[i] - interp_yield
        
# Set curvature for first and last maturities to NaN (since they have no neighbors)
curvatures[0] = np.nan
curvatures[-1] = np.nan

curvature_df = pd.DataFrame({
    'Maturity (Years)': df.columns[1:],
    'Curvature': curvatures
})

curvature_df
```
